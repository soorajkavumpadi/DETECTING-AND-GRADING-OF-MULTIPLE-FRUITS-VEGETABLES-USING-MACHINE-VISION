```python
import cv2
```

```python
import numpy as np
import os
```

```python
cap = cv2.VideoCapture('apple-timelapse video/1.mp4')
```

```python
try:
    if not os.path.exists('datafromvideo'):
        os.makedirs('datafromvideo')
except OSError:
    print ('Error: Creating directory of data')
```

```python
currentFrame = 0
K=0
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Saves image of the current frame in jpg file
    
    if K%500 == 0:
        name = './data2/frame' + str(currentFrame) + '.jpg'
        print ('Creating...' + name)
        cv2.imwrite(name, frame)
    K=K+1
        
        
    

    # To stop duplicate images
    currentFrame += 1

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
